**Tokenization**

```python
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')  # Download punkt

# remove the index and square brackets
fin3['Sentence']  # Your list of sentences

# Tokenize each sentence in the list 'a'
tokenized_sentences = [word_tokenize(sentence) for sentence in fin3['Sentence']]

# Print the first few tokenized sentences to verify
print(tokenized_sentences[:5])
```

**Stemming**

```python
#Poster stemmer
import nltk
from nltk.stem.porter import PorterStemmer
nltk.download('punkt')

# Assuming 'tokenized_sentences' is your list of tokenized sentences
porter = PorterStemmer()

stemmed_sentences = []
for sentence in tokenized_sentences:
    stemmed_sentence = []
    for word in sentence:
        stemmed_word = porter.stem(word)
        stemmed_sentence.append(stemmed_word)
    stemmed_sentences.append(stemmed_sentence)

# Print the first few stemmed sentences to verify
print(stemmed_sentences[:5])
```

```python
#Lancasterstemmer
import nltk
from nltk.stem.lancaster import LancasterStemmer
nltk.download('punkt')

# Assuming 'tokenized_sentences' is your list of tokenized sentences
lancaster = LancasterStemmer()

stemmed_sentences = []
for sentence in tokenized_sentences:
    stemmed_sentence = []
    for word in sentence:
        stemmed_word = lancaster.stem(word)
        stemmed_sentence.append(stemmed_word)
    stemmed_sentences.append(stemmed_sentence)

# Print the first few stemmed sentences to verify
print(stemmed_sentences[:5])
```

```python
#using showball
import nltk
nltk.download("showball_data")
from nltk.stem.snowball import SnowballStemmer
nltk.download('punkt')
snowball = SnowballStemmer(language='english')
stemmed_sentences = []
for sentence in tokenized_sentences:
    stemmed_sentence = []
    for word in sentence:
        stemmed_word = snowball.stem(word)
        stemmed_sentence.append(stemmed_word)
    stemmed_sentences.append(stemmed_sentence)

# Print the first few stemmed sentences to verify
print(stemmed_sentences[:5])
```

**Lematization**

```python
#Lemmatization
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')

# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

# Tokenize and lemmatize each sentence in the list 'a'
lemmatized_sentences = [[lemmatizer.lemmatize(token.lower(),pos='n') for token in word_tokenize(sentence)] for sentence in fin3['Sentence']]

# Print the first few lemmatized sentences to verify
print(lemmatized_sentences[:5000])
```

**Stopwords**

```python

import spacy

nlp = spacy.load("en_core_web_sm")

filtered_lemmatized_sentences = []
for sentence in lemmatized_sentences:
    filtered_sentence = []
    for word in sentence:
        doc = nlp(word.lower())  # Process each word with spaCy, converting to lowercase
        if not doc[0].is_stop and doc[0].is_alpha:  # Check if the word is a stop word or punctuation
            filtered_sentence.append(doc[0].text)  # Append the lowercase, filtered word
    filtered_lemmatized_sentences.append(filtered_sentence)

# Print the first few filtered, lemmatized sentences to verify
print(filtered_lemmatized_sentences[:5])
```

Remove html tags and Ner

```python
import spacy
from bs4 import BeautifulSoup

nlp = spacy.load("en_core_web_sm")

# Assuming 'filtered_lemmatized_sentences' contains the output of your stop word removal
final_sentences = []
for sentence in filtered_lemmatized_sentences:
    # HTML Removal
    sentence_text = " ".join(sentence)  # Join the words back into a string
    soup = BeautifulSoup(sentence_text, "html.parser")
    cleaned_text = soup.get_text()

    # NER Removal
    doc = nlp(cleaned_text)
    filtered_tokens = [token.text for token in doc if not token.ent_type_]
    final_sentences.append(filtered_tokens)

# Print the first few final sentences to verify
print(final_sentences[:5])
```

Word2vec

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from gensim.models import Word2Vec
import numpy as np

# Extract features and target
sentences = fin3['Sentence']
labels = fin3['Sentiment']

# Encode the target labels
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(labels)

# Preprocess text: Tokenize sentences
tokenized_sentences = [sentence.split() for sentence in sentences]

# Train a Word2Vec model on the tokenized sentences
word2vec_model = Word2Vec(sentences=tokenized_sentences, vector_size=100, window=5, min_count=1, workers=4)

# Function to get sentence vectors by averaging word vectors
def sentence_to_vector(sentence, model, vector_size):
    words = sentence.split()
    word_vectors = [model.wv[word] for word in words if word in model.wv]
    if word_vectors:
        return np.mean(word_vectors, axis=0)
    else:
        return np.zeros(vector_size)

# Convert all sentences to vectors
vectorized_sentences = np.array([sentence_to_vector(sentence, word2vec_model, 100) for sentence in sentences])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(vectorized_sentences, encoded_labels, test_size=0.2, random_state=42)

# Summary of the process
X_train.shape, X_test.shape, len(y_train), len(y_test)
```

```python
!pip install imblearn
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from imblearn.over_sampling import SMOTE
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split

# Assuming 'final_sentences' and 'fin' DataFrame are available from previous steps

# 1. Create a DataFrame with processed sentences and sentiment labels
fin_processed = pd.DataFrame({'Processed_Sentence': [' '.join(sentence) for sentence in final_sentences],
                              'Sentiment': fin3['Sentiment']})

# 2. Initialize LabelEncoder and encode sentiment labels
label_encoder = LabelEncoder()
label_encoder.fit(fin_processed['Sentiment'])
fin_processed['Encoded_Sentiment'] = label_encoder.transform(fin_processed['Sentiment'])

# 3. TF-IDF Vectorization
tfidf = TfidfVectorizer(max_features=5000)
X = tfidf.fit_transform(fin_processed['Processed_Sentence']).toarray()
y = fin_processed['Encoded_Sentiment']

# 4. Handle Class Imbalance with SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# 5. Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)

# 6. Train the Logistic Regression model
classifier = LogisticRegression(max_iter=10000, C=100, random_state=42)
classifier.fit(X_train, y_train)

# 7. Predict on the test set
y_pred = classifier.predict(X_test)

# 8. Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, target_names=label_encoder.classes_)

# 9. Print the results
print(f"Accuracy: {accuracy:.2f}")
print("\nClassification Report:\n")
print(report)
```
